```scala211
import scala.collection.mutable.ListBuffer // Might need this later
import scala.io.Source

val filename = "input"
val lines = Source.fromFile(filename).getLines.toList
```

```scala211
// Separately parse initial state and rules
var offset = 250 // This means plant 20 is actually plant 0 (Might not matter)
var state = "."*offset + lines(0).split(':')(1).trim + "."*offset
var newState = state.toCharArray
val rules = lines.slice(2, lines.length).map(x => x.split(" => ")).map(x => x(0) -> x(1).charAt(0)).toMap 

// Initialise offset from 0 - this can only ever grow to 20 either side so just start with this
var totalCount = state.filter(x => x == '#').length
val numSeconds = 200
```

```scala211
//state.foreach(print)
//println()
for (second <- 1 to numSeconds) {
    println("Second: ", second)
    newState = ("."*state.length).toCharArray
    for (i <- 2 until state.length-2) {
        val neighbours = state.slice(i-2, i+3)
        //println(neighbours)
        //println(rules(neighbours))
        if(rules.contains(neighbours)) {
            newState(i) = rules(neighbours)
        }
        else {
            newState(i) = '.'
        }
    }
    state.foreach(print)
    println()
    newState.foreach(print)
    println()
    totalCount += state.filter(x => x == '#').length
    val prevIndexedState = state.zipWithIndex.map(x => (x._1, x._2-offset))
    val prevPotSum = prevIndexedState.filter(x => x._1 == '#').map(x=> x._2).sum
    state = newState.mkString
    println("Pots width: ", state.lastIndexOf('#') - state.indexOf('#'))
    //println()
    //println(totalCount)
    val indexedState = state.zipWithIndex.map(x => (x._1, x._2-offset))
    val potSum = indexedState.filter(x => x._1 == '#').map(x=> x._2).sum
    //indexedState.foreach(x => print(x._2.toString.padTo(3, ' ')))
    //println()
    //indexedState.foreach(x => print(x._1.toString.padTo(3, ' ')))
    println(prevPotSum, potSum, potSum-prevPotSum)
    println()
}
```
